<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
df["SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT"] = df["SUPPORT_SERVICES_EXPENDITURE"] / df["GRADES_8_G"]
df["PROPORTION_TOTAL_EXPENDITURE_ON_INSTRUCTION"] = df["INSTRUCTION_EXPENDITURE"] / df["TOTAL_EXPENDITURE"]
```

Feature engineering justification: More or less money spent on support services per student may affect their math performance. A greater proportion of a state's total expenditure on instruction may mean that the state puts more emphasis on education, and vice versa, which would impact students' scores.

<h2>Visualization</h2>

Investigate the relationship between your chosen response variable and at least two predictors using visualizations. Write down your observations.

**Visualization 1**

```python
df.plot.scatter(x = "SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT", y = "AVG_MATH_8_SCORE", alpha = 0.6)
plt.xlabel("Support Services Expenditure per Student")
plt.ylabel("8th Grade Math Score") 
```

There seems to be a slight positive correlation between support services expenditure per student and students' 8th grade math score. The data is concentrated towards the lower end of the x-axis, suggesting that states spend roughly the same amount on support for students. 

**Visualization 2**

```python
df.plot.scatter(x = "AVG_MATH_4_SCORE", y = "AVG_MATH_8_SCORE", alpha = 0.6)
plt.xlabel("4th Grade Math Score")
plt.ylabel("8th Grade Math Score")
```

There is a strong positive correlation between 4th grade math scores and 8th grade math scores, suggesting that how students perform in lower grade levels may predict how well they tend to do in higher grade levels.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[["SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT","AVG_MATH_4_SCORE"]].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=71)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.score(X_test, y_test))
np.mean(np.abs(model.predict(X_test)-y_test))
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = "AVG_MATH_4_SCORE"

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel("4th Grade Math Score")
plt.title("Model Behavior On Training Set")
```

```python
col_name = "AVG_MATH_4_SCORE"

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel("4th Grade Math Score")
plt.title("Model Behavior on Testing Set")
```
